```python
import pandas as pd
import datashader as ds
import datashader.transfer_functions as tf
from colorcet import fire,kbc,bmw,gray
import plotly.express as px

```

```python
df=pd.read_csv("/content/Crimes_-_2001_to_Present.csv")
df.head()
```

```python
df.shape
```

```python
dff=df[['Primary Type','Latitude','Longitude']]
dff['Primary Type'].unique()
```

```python
dff=dff[dff['Primary Type'].isin(['PROSTITUTION'])]
```

```python
dff
```

```python
dff.dropna(subset=['Latitude','Longitude'],inplace=True)
dff
```

```python
dff.shape
```

```python
csv=ds.Canvas(plot_width=500,plot_height=500)
csv
```

```python
aggs=csv.points(dff,x='Longitude',y='Latitude')
```

```python
aggs
```

```python
coords_lat,coords_lon=aggs.coords['Latitude'].values,aggs.coords['Longitude'].values
```

```python
from xarray.core import coordinates
coordinates=[
    [coords_lon[0],coords_lat[0]],
    [coords_lon[-1],coords_lat[0]],
    [coords_lon[-1],coords_lat[-1]],
    [coords_lon[0],coords_lat[-1]]
]
```

```python
img=tf.shade(aggs,cmap=bmw,how='eq_hist',alpha=255)[::-1].to_pil()
```

```python
fig=px.scatter_mapbox(dff[:1],lat='Latitude',lon='Longitude',zoom=10)
```

```python

fig.update_layout(mapbox_style='carto-darkmatter',
                  mapbox_layers=[
                      {
                          'sourcetype':'image',
                           'source':img,
                          'coordinates':coordinates,
                            

                      }
                  ])
fig.show()
```
